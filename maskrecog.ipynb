```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from keras.preprocessing import image
import os
import shutil 
import cv2
from imutils import paths
from keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
```

```python

data = []
label = []

folder_path1 = "with_mask"
for i in os.listdir(folder_path1):
    if not i.startswith('.') and i.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")):
        image_path = os.path.join(folder_path1, i)
        image = load_img(image_path, target_size=(224,224))
        image = img_to_array(image)
        data.append(image)
        label.append("with_mask")

folder_path2 = "without_mask"
for i in os.listdir(folder_path2):
    if not i.startswith('.') and i.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")):
        image_path = os.path.join(folder_path2, i)
        image = load_img(image_path, target_size=(224,224))
        image = img_to_array(image)
        data.append(image)
        label.append("without_mask")
```

```python
from tensorflow.keras.layers import Dense
model.add(Dense(224,activation = "relu"))
model.add(Dense(112,activation = "relu"))
model.add(Dense(56,activation = "relu"))
model.add(Dense(1,activation='sigmoid'))
```

```python

def detect_face_mask(img):
    probabilities = model.predict(img.reshape(1,224,224,3)) 
    if probabilities > 0.5:
        return 0  
    else:
        return 1  
```

```python
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')


while True:
    ret, frame = cap.read()
    
    faces = face_cascade.detectMultiScale( frame ,minSize=(30, 30))


    
    for (x, y, w, h) in faces:
        
        face_image = frame[y:y+h, x:x+w]
        face_image = cv2.resize(face_image, (224, 224))
        
        prediction = detect_face_mask(face_image);
        font = cv2.FONT_HERSHEY_SIMPLEX
        if(prediction == 1):
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0))
            cv2.putText(frame, "mask detected", (x, y-10),font,0.7, (0, 255, 0))
        else:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255))
            cv2.putText(frame, "mask not detected", (x, y-10),font,0.7, (0,0,255))
    
    
    cv2.imshow('Window', frame)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cap.release()
cv2.destroyAllWindows()
```
